```python
### Part # 1:
### Stage 0: Here we will Import the necessary modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Stage 1: Get the data from yahoo finance 
### on the daily closing prices of Tesla, GM and Ford.
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2020,8,16)
```

```python
### Tesla Stock (Ticker: TSLA on the NASDAQ)
tesla = web.DataReader("TSLA", 'yahoo',start,end)
### Ford Stock (Ticker: FORD on the NASDAQ)
ford = web.DataReader("F", 'yahoo',start,end)
### GM Stock (Ticker: General Motors on the NASDAQ)
gm = web.DataReader("GM", 'yahoo',start,end)
```

```python
### Stage 2: Savings data to csv files
tesla.to_csv('Tesla_Stock.csv')
ford.to_csv('Ford_Stock.csv')
gm.to_csv('GM_Stock.csv')
```

```python
### Part # 2: Visualise the data, and visually inspect interesting events.
### In this section, we also create a candlestick chart from scratch.

# Plot on opening prices
tesla["Open"].plot(figsize=(15,7))
gm["Open"].plot()
ford["Open"].plot()
plt.ylabel("Stock Price")
plt.xlabel("Stock price of Tesla, GM and Ford")
plt.legend()
```

```python
# Plot on daily Volumes
tesla["Volume"].plot(label="Tesla",figsize=(15,7))
gm["Volume"].plot(label="GM")
ford["Volume"].plot(label="Ford")
plt.ylabel("Volume Traded")
plt.legend()
```

```python
ford.iloc[[ford["Volume"].argmax()]]
```

```python
ford.iloc[470:520]["Open"].plot()
```

```python
# Dollars traded for the day
tesla["Total Traded"] = tesla["Open"] * tesla["Volume"]
gm["Total Traded"] = gm["Open"] * gm["Volume"]
ford["Total Traded"] = ford["Open"] * ford["Volume"]
```

```python
# Plotting moving average of GM
gm["Open"].plot(label="No Moving average",figsize=(15,7))
gm["MA50"] = gm["Open"].rolling(50).mean()
gm["MA50"].plot(label="MA50")
gm["MA200"] = gm["Open"].rolling(200).mean()
gm["MA200"].plot(label="MA200")
plt.legend()
```
